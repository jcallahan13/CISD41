**Exercise 1 (b)** (worth 15 points)

** Follow along with these steps:**
- Create a figure object called fig 
- Add an axes (named axes1) to the figure canvas at [0,0,0.8,0.8]. 
- Plot (x,y) on that axes axes1
- Set the labels
- Set the title to match the plot below:

```python
print ("This is done by John Callahan")
fig=plt.figure()    
axes1=fig.add_axes([0,0,0.8,0.8])
axes1.plot(x,y)
axes1.set_xlabel('X values')
axes1.set_ylabel('Y values')
axes1.set_title('X vs Y')
    
```

**Exercise 4 (b) (Worth 10 points)**
- Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
print ("This is done by John Callahan")
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,'g',lw=3,ls='--')
axes[1].plot(x,z,'b',lw=5,ls='-.')
```
